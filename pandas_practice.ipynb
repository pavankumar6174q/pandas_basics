```python
#creating a data frame using pandas
import pandas as pd
d1 = {"id":[201,202,203,204,205],
     "gender":['m','f','m','f','m'],
     'age':[25,26,23,24,23]}
df2 = pd.DataFrame(d1)
df2
```

```python
d2 = {"id":[206,207,208,209,210],
     "gender":['m','f','m','f','m'],
     'age':[20,29,65,32,29]}
df1 = pd.DataFrame(d2)
df1
```

```python
#concatinating
df3 = pd.concat([df1,df2],ignore_index = True)
df3
```
